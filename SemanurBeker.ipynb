```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('CountryHits240901.csv')
spotify = veriSeti['Track']
dinlenme_says = veriSeti['Tempo']

plt.figure(figsize=(20, 7))  # Grafiin boyutunu ayarlayn
plt.bar(spotify, dinlenme_says)
plt.xlabel('spotify')
plt.ylabel('Dinlenme Says')
plt.title('Spotifyda lkelere Gre Mzik Sralamas')
plt.xticks(rotation=90)  # lke isimlerinin dikey yazlmas
plt.show()
```
