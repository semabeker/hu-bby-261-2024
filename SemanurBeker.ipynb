```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('CountryHits240901.csv')
spotify = veriSeti['Track']
dinlenme_sayısı = veriSeti['Tempo']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(spotify, dinlenme_sayısı)
plt.xlabel('spotify')
plt.ylabel('Dinlenme Sayısı')
plt.title('Spotifyda Ülkelere Göre Müzik Sıralaması')
plt.xticks(rotation=90)  # Ülke isimlerinin dikey yazılması
plt.show()
```
